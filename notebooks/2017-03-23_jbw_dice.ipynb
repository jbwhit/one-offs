```python
%matplotlib inline
%config InlineBackend.figure_format='retina'
# %load_ext autoreload
# the "1" means: always reload modules marked with "%aimport"
# %autoreload 1

from __future__ import absolute_import, division, print_function
import matplotlib as mpl
from matplotlib import pyplot as plt
from matplotlib.pyplot import GridSpec
import seaborn as sns
import mpld3
import numpy as np
import pandas as pd
import os, sys
import warnings
```

```python
def roll():
    return np.random.choice([1,2,3,4,5,6])
```

```python
def play_a_round(threshold=3.5):
    cash = roll()
    if cash < threshold:
        cash = roll()
    return cash
```

```python
thresholds = []
exp_values = []
for threshold in [.5, 1.5, 2.5, 3.5, 4.5, 5.5, 6.5]:
    thresholds.append(threshold)
    exp_value = np.average([play_a_round(threshold=threshold) for _ in range(10000)])
    print(threshold, exp_value)
    exp_values.append(exp_value)
```

```python
sns.set_context('poster', font_scale=1.3)
```

```python
fig, ax = plt.subplots(figsize=(12, 8))
ax.scatter(thresholds, exp_values)
ax.set_ylabel("Expected Value")
ax.set_xlabel("Threshold Value")
fig.tight_layout();
```
